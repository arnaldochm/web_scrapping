```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from unicodedata import normalize
import re
import requests
```

```python
url = "https://en.wikipedia.org/wiki/Firearm_death_rates_in_the_United_States_by_state"
response = requests.get(url).text

soup = BeautifulSoup(response, 'html.parser')
```

```python
#Get all the tables
tables  = soup.find_all('table')
tables
```

```python
i = 1

for table in tables:
    if i == 1:
        headers = [header.text.strip() for header in table.find_all('th')]
        rows = []
        table_rows = table.find_all('tr')
        for row in table_rows:
            td = row.find_all('td')
            row = [row.text for row in td]
            rows.append(row)
    i += 1
```

```python
rows
```

```python
headers
headers.append('blank')
```

```python
df_raw = pd.DataFrame(rows, columns=headers)
df_raw
```

```python
df = df_raw.copy()

df = df.drop(['blank'], axis = 1)

df = df.drop(df.index[:1], axis = 0)

df["State"] = df["State"].str.normalize('NFKD')

df['State'] = df['State'].str.replace(' *', '')

df
```

```python
j = 1

for table in tables:
    if j == 2:
        headers = [header.text.strip() for header in table.find_all('th')]
        rows = []
        table_rows = table.find_all('tr')

        for row in table_rows:
            td = row.find_all('td')
            cell = [cell.text for cell in td]
            rows.append(cell)
    j += 1
```

```python
clean_headers = [re.sub('\n','', x) for x in headers]

#Remover lo que haya dentro de los parentesis
clean_headers = [re.sub('\(.*\)','', x) for x in clean_headers]

#Limpiar espacios en blanco
clean_headers = [x.strip() for x in clean_headers]

clean_headers = [x for x in clean_headers if x != '']

clean_headers
```
